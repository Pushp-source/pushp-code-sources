```
import csv
import pandas as pd
import matplotlib.pyplot as plt
import datetime

col_list = ["date","China","India", "Italy","Iran", "Japan", "South_Korea", "United_Kingdom"]
df = pd.read_csv('total_cases.csv',usecols=col_list,encoding='latin-1')
#print(df.China)
date_india = []
date_china =[]
date_italy = []
date_iran =[]
date_japan =[]
date_skorea =[]
date_uk =[]
cases_india = []
cases_china =[]
cases_italy =[]
cases_iran = []
cases_japan =[]
cases_skorea =[]
cases_uk =[]
idx = 0

for x in df.China:  
  if(x>100):
    idx = idx +1
    #print(x,'',df.date[idx])
    date_china.append(idx)
    cases_china.append(x)
idx = 0
for x in df.India:  
  if(x>100):
    idx = idx +1
    #print(x,'',df.date[idx])
    date_india.append(idx)
    cases_india.append(x)
idx = 0
for x in df.Italy:  
  if(x>100):
    idx = idx +1
    #print(x,'',df.date[idx])
    date_italy.append(idx)
    cases_italy.append(x)
idx = 0
for x in df.Iran:  
  if(x>100):
    idx = idx +1
    #print(x,'',df.date[idx])
    date_iran.append(idx)
    cases_iran.append(x) 
idx = 0
for x in df.Japan:  
  if(x>100):
    idx = idx +1
    #print(x,'',df.date[idx])
    date_japan.append(idx)
    cases_japan.append(x)
idx = 0
for x in df.South_Korea:  
  if(x>100):
    idx = idx +1
    #print(x,'',df.date[idx])
    date_skorea.append(idx)
    cases_skorea.append(x)          
idx = 0
for x in df.United_Kingdom:  
  if(x>100):
    idx = idx +1
    #print(x,'',df.date[idx])
    date_uk.append(idx)
    cases_uk.append(x) 
    
plt.semilogy(date_india,cases_india,label ="india",c='red',marker='o')
plt.semilogy(date_china,cases_china,label = "china",c='blue',marker='o')
plt.semilogy(date_italy,cases_italy,label = "italy",c='green',marker='o')
plt.semilogy(date_iran,cases_iran,label = "iran",c='yellow',marker='o')
plt.semilogy(date_japan,cases_japan,label = "japan",c='pink',marker='o')
plt.semilogy(date_skorea,cases_skorea,label = "south korea",c='brown',marker='o')
plt.semilogy(date_uk,cases_uk,label = "united kingdom",c='black',marker='o')
plt.title("Plotting cases and deaths of corona")
plt.xlabel('days since passed 100 case')
plt.ylabel('deaths')
plt.grid()
plt.legend()
plt.show()

   
```
